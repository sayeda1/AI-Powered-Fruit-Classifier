```python
# Importing Libraries
import pandas as pd
data = pd.read_csv("data\\Transactions.csv")
```

```python
# Trim Dataset
data = data[data['trans_group_en'].str.lower() != 'gifts'] # Remove all properties that were gifts/grants
print(data['instance_date'].dtype)
#data = data[data['instance_date'] <= '25-05-25']

data.head(n = 10)
```

```python
# Aggregate Data By "area_name_en"
area_stats = data.groupby('area_name_en').agg({
'actual_worth': ['mean', 'median'],
'transaction_id': 'count',
'procedure_area': 'mean'
}).reset_index()

area_stats.columns = ['area','avg_price','median_price','num_transactions','avg_area']

area_stats.head()
```

```python
# Preprocess Data - Normalize Before Clustering
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_area_stats = scaler.fit_transform(area_stats[['avg_price','median_price','num_transactions','avg_area']])
```

```python
# Perform Clustering
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
area_stats['cluster'] = kmeans.fit_predict(scaled_area_stats)
```

```python
# Visualize Data
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=area_stats, x='cluster')
plt.title("Number of Areas per Cluster")
plt.xlabel("Cluster")
plt.ylabel("Count")
plt.show()
```

```python
sns.scatterplot(
    data=area_stats,
    x='avg_area',
    y='avg_price',
    hue='cluster',
    palette='Set2',
    s=100
)
plt.title("Clusters of Dubai Areas")
plt.xlabel("Average Area (sqm)")
plt.ylabel("Average Price (AED)")
plt.show()
```
